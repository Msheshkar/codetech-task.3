```python
# TensorFlow – neural network aur pre-trained model ke liye
import tensorflow as tf

# NumPy – numerical operations ke liye
import numpy as np

# Matplotlib – images display karne ke liye
import matplotlib.pyplot as plt

# PIL – image load aur resize karne ke liye
from PIL import Image
```

```python
def load_img(path):
    # Image ko open karo
    img = Image.open(path)

    # Image ka size fix kar rahe hain (fast processing ke liye)
    img = img.resize((512, 512))

    # Image ko array me convert karo
    img = np.array(img)

    # Model ke input ke liye extra dimension add karo
    img = tf.expand_dims(img, axis=0)

    return img
```

```python
# Normal photo (content)
content_image = load_img("my pic.jpg")

# Artistic painting (style)
style_image = load_img("download.jpeg")
```

```python
# VGG19 ek famous CNN hai jo ImageNet par trained hai
vgg = tf.keras.applications.VGG19(
    include_top=False,   # Fully connected layers remove
    weights='imagenet'   # Pre-trained weights use
)

# Model ke weights freeze kar rahe hain
vgg.trainable = False
```

```python
# Content image ko VGG model ke format me convert kar rahe hain
stylized_image = tf.keras.applications.vgg19.preprocess_input(content_image)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.title("Content Image")
plt.imshow(content_image[0].numpy().astype('uint8'))
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Style Image")
plt.imshow(style_image[0].numpy().astype('uint8'))
plt.axis('off')

plt.show()
```
